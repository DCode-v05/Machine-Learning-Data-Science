```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
fig , ax = plt.subplots(figsize=(7,7))
ax.plot(x,y)
ax.set(title="Square" , xlabel="No." , ylabel="Squ No.")
plt.show()
fig.savefig("sample.png")
```

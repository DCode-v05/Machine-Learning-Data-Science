```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
car=pd.read_csv("car-sales.csv")
car
```

```python
car["Price"] = car["Price"].str.replace('[\$\,\.]' ,'' ,regex=True).astype(int)
car
```

```python
car["Price"]=car["Price"]//100
car
```

```python
car["Sales Date"]=pd.date_range("1/1/2024",periods=len(car))
car
```

```python
car["Total Price"] = car["Price"].cumsum()
car
```

```python
car.plot("Sales Date" , "Total Price");
```

```python
car.plot("Odometer (KM)" , "Price" , kind="scatter");
```

```python
car.plot("Odometer (KM)" , "Price" , kind="bar");
```

```python
car.plot("Odometer (KM)" , "Price" , kind="hist");
```

```python
heart=pd.read_csv("heart-disease.csv")
heart.head()
```

```python
heart.plot.hist(subplots=True,figsize=(10,30));
```

```python
over50=heart[heart["age"]>50]
over50.head()
```

```python
fig , ax =plt.subplots(figsize=(10,6))
over50.plot(kind="scatter" , x="age" , y="chol" , c="thalach" , ax=ax);
ax.set_xlim([45,80]);
```

```python
plt.style.available
```

```python
plt.style.use("seaborn-v0_8-whitegrid")

fig , (ax0 , ax1) = plt.subplots(nrows=2 , ncols=1 , figsize=(10,10) , sharex=True)
sca1=ax0.scatter(x=over50["age"] , y=over50["chol"] , c=over50["target"] , cmap="winter");
ax0.set(ylabel="Cholestral" , title="Age Vs Cholestral");
ax0.axhline(over50["chol"].mean() , ls="--" , color="Red")
ax0.legend(*sca1.legend_elements() , title="Target");
ax0.set_xlim([50,80]);

sca2=ax1.scatter(x=over50["age"] , y=over50["thalach"] , c=over50["target"] , cmap="winter");
ax1.set(xlabel="Age" , ylabel="Heart Rate" , title="Age Vs Heart Heart");
ax1.axhline(over50["thalach"].mean() , ls="--" , color="Red")
ax1.legend(*sca2.legend_elements() , title="Target");
ax1.set_xlim([50,80]);
ax1.set_ylim([60,200]);

fig.suptitle("Heart Disease Analysis" , fontsize=16 , fontweight="bold");
```

```python
fig.savefig("Heart Disease.png")
```
